```python
# Univeriate Annalysis
# Count plot
sns.countplot(x = 'gender', data = df)
plt.xlabel('Gender')
plt.ylabel('state')
plt.title('Gender Distribution')
```

```python
# Replace with REAL column names from df.columns
value_col = df.select_dtypes(include="number").columns[0]   # first numeric column
group_col = df.select_dtypes(include="object").columns[0]   # first group column
```

```python
# Distribution Comparison
plt.figure(figsize=(6,4))
sns.histplot(data=df, x=value_col, hue=group_col, kde=True)
plt.title("Distribution Comparison Across Groups")
plt.show()
```

```python
# Boxplot Comparison
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x=group_col, y=value_col)
plt.title("Spread & Outliers by Group")
plt.show()
```

```python
# Select numeric columns
num_df = df.select_dtypes(include="number")

# Histograms + KDE
for col in num_df.columns:
    plt.figure()
    sns.histplot(num_df[col], kde=True)
    plt.title(f"Histogram & KDE: {col}")
    plt.show()
```

```python
# Select numeric columns
num_df = df.select_dtypes(include="number")

#  Boxplots
for col in num_df.columns:
    plt.figure()
    sns.boxplot(x=num_df[col])
    plt.title(f"Boxplot: {col}")
    plt.show()
```

```python
# Correlation Heatmap if there are four or more numeric columns
numeric_df = df.select_dtypes(include='number')
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10,8))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.show()
```

```python
# Select numeric columns
num_df = df.select_dtypes(include="number")

# Outlier Detection (IQR method)
Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

outliers = (num_df < (Q1 - 1.5 * IQR)) | (num_df > (Q3 + 1.5 * IQR))
print("Outliers count:")
print(outliers.sum())

# Skewness & Spread
print("\nSkewness:")
print(num_df.skew())

print("\nSpread (Standard Deviation):")
print(num_df.std())

# Simple Visualization
for col in num_df.columns:
    plt.figure()
    sns.boxplot(x=num_df[col])
    plt.title(col)
    plt.show()
```

```python
import os

# Select numeric columns
num_df = df.select_dtypes(include="number")

# Create folder
os.makedirs("output", exist_ok=True)

# Save Plots
for col in num_df.columns:
    plt.figure()
    plt.boxplot(num_df[col])
    plt.title(col)
    plt.savefig(f"output/{col}_boxplot.png")
    plt.close()

# Save Interpretation
text = """
The data shows some outliers in numeric columns, visible from boxplots.
Some variables are right-skewed, meaning most values are low with few high values.
The spread of data varies across features, indicating different levels of
variability. Outliers may impact model performance, so data cleaning or
normalization is recommended before applying machine learning models.
"""

with open("output/interpretation.txt", "w") as f:
    f.write(text)

print("Plots and interpretation saved.")
```
